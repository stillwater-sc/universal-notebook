# Decimal integers: decimal

Adaptive-size big integers are typically required for integer operations that involve factorials, binomials, and integer sequences. The _decimal_ class in _Universal_ provides a decimal adaptive integer for these types of studies.

First import the global header file `universal.hpp` to get the notebook setup.

```c++
#include "universal.hpp"
```

Let's implement a simple kernel that computes the product of two numbers

Include the top-level header file of the `decimal` data type to gain access to the number system.

```c++
#include <universal/number/decimal/decimal.hpp>
```

Some basic decimal math:

```c++
using Decimal = sw::universal::decimal;
Decimal a, b, c;
a = 10;
b = 10;
c = ipow(10, 10);
std::cout << "10^10 = " << c << std::endl;
```
